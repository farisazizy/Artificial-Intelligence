```
#Import library yang dibutuhkan

import math
import random
```

```
#Perhitungan Fitness

def hitungFitness(x, y):
  return math.cos(x) * math.sin(y) - (x / (y * y + 1))
```

```
def genpop(npop):
  pop = [[random.uniform(-1.0, 2.0), random.uniform(-1.0, 1.0)] for j in range(npop)]
  return pop
```

```
def randomparent(npop):
    return random.randint(0, npop)
```

```
def crossover(a1, a2):
    rand = random.random()
    titik = 1
    if rand <= pCross:
        for k in range(titik):
            a1[k], a2[k] = a2[k], a1[k]
    return a1, a2
```

```
def mutation(a):
    rand = random.random()
    titik = random.randint(0, nkrom - 1)
    if rand <= pmutasi:
      a[titik] += random.uniform(-1, 1)
        # if a[titik] == 0:
        #   a[titik] = 1
        # else:
        #   a[titik] = 0
    return a
```

```
def sort_fitness(fitness):
    return sorted(range(len(fitness)), key=lambda k: fitness[k], reverse=True)
```

```
def Steadystate(gab): 
    fitness = []
    for f in range(len(gab)):
        fitness.append(hitungFitness(gab[f][0], gab[f][1]))
    steadystate_list = sort_fitness(fitness)
    pop = []
    for j in range(nPop):
        pop.append(gab[steadystate_list[j]])

    return pop, fitness
```

```
nGen = 100
pCross = 0.8
pmutasi = 0.1
nPop = 2000
nkrom = 2
pop = genpop(nPop, nkrom)
        
for i in range(nGen):

    # fitness = []
    anak = []

    for j in range(nPop // 2):

        tournament = []

        z = 0
        while z < 5:
          tournament.append(random.randint(0, nPop-1))
          z += 1

        # print(tournament)

        parent1 = randomparent(len(tournament))
        parent2 = randomparent(len(tournament))

        # print(parent1)

        anak1 = pop[parent1][:]
        anak2 = pop[parent2][:]

        anak1, anak2 = crossover(anak1, anak2)
        anak1 = mutation(anak1)
        anak2 = mutation(anak2)

        anak.append(anak1)
        anak.append(anak2)

    gab = pop + anak

    pop, fitness = Steadystate(gab)
    print('gen', i+1)
    print('x: ', pop[0][0])
    print('y: ', pop[0][1])
    print('mean: ', sum(fitness)/len(fitness))
    print('fitness: ', fitness[0], '\n')

# result = pop[0]
# fitness = fitness[0]
```
